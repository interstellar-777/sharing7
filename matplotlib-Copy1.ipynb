```python
#With Pyplot, you can use the pie() function to draw pie charts:
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 25, 25, 15])
mylabels=['apples','bananas','cherries','grapes']
plt.pie(y,labels=mylabels) #add labels to the pie chart with the label parameter.The labels parameter must be an array with one label for each wedge:
plt.show()
print('\n')
plt.pie(y)
plt.show()


#he size of each wedge is determined by comparing the value with all the other values, by using this formula:
#The value divided by the sum of all values: x/sum(x)

#As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.
#The startangle parameter is defined with an angle in degrees, default angle is 0:
plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
print('\n')


myexplode = [0.2, 0, 0, 0]  #Pulls the "Apples" wedge 0.2 from the center of the pie:
plt.pie(y, labels = mylabels, explode = myexplode)
plt.show()
print('\n')

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True) #adds a shadow to pie chart
plt.show()
print('\n')

mycolors = ["black", "hotpink", "blue", "orange"]
plt.pie(y, labels = mylabels, colors = mycolors,explode = myexplode, shadow = True)
plt.show()
print('\n')
#'r' - Red 'g' - Green 'b' - Blue 'c' - Cyan 'm' - Magenta 'y' - Yellow 'k' - Black 'w' - White


plt.pie(y, labels = mylabels)
plt.legend()   #adds legend to the piechart
plt.show() 
print('\n')


plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
ax.plot_surface(X, Y, Z)
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("3D Plot")
plt.show()
```
