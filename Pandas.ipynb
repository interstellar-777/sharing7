```python
#map()
#syntax- series.map(arg,na_action=None) or na_action="ignore". by default na_action=none
import numpy as np
import pandas as pd
a=pd.Series(["java","c++","cpro",np.nan,"python"])
print(a)
b=a.map({"java":"core"})
print(b)
c=a.map({"java":"core","c++":"c","cpro":"ruby",})
print(c)
d=a.map('I am a {}'.format)
print(d)
print("\n")
e=a.map('I am a {}'.format,na_action='ignore')
print(e)
```

```python
#dataframe.join(other, on, how, lsuffix, rsuffix, sort)  #how-specifies which index to use i.e left,right,outer,inner default left
import pandas as pd
data1 = {"name": ["Sally", "Mary", "John"],"age": [50, 40, 30]}
data2 = {"qualified": [True, False, False]}
df1 = pd.DataFrame(data1,index=['a','b','c'])
df2 = pd.DataFrame(data2)
newdf = df1.join(df2) #default left join
print(newdf)
print('\n')
newdf1 = df1.join(df2,how='left')
print(newdf1)
print('\n')
newdf2 = df1.join(df2,how='right')
print(newdf2)
print('\n')
newdf4 = df2.join(df1)
print(newdf4)
```

```python
#merge example we can merge two dataframes when the have one column common and other columnnames different
import pandas as pd 
data1 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32],} 
data2 = {'key': ['K0', 'K1', 'K2', 'K3'],
         'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
df = pd.DataFrame(data1)
df1 = pd.DataFrame(data2) 
print(df, "\n\n", df1) 
print('\n')
res = pd.merge(df, df1)
print(res)
print('\n')
res1 = pd.merge(df1, df)
print(res1)
```
